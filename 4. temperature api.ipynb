```python
import requests
import json
import datetime

today = datetime.date.today()
year = today.year
year_prev = year - 1

# Variables for in the url
latitude = 52.01
longitude = 4.36
start_date = f"{year_prev}-01-01"
end_date = f"{year_prev}-12-31"

# climate url
url = f"https://climate-api.open-meteo.com/v1/climate?latitude={latitude}&longitude={longitude}&start_date={start_date}&end_date={end_date}&daily=temperature_2m_max,windspeed_10m_max,relative_humidity_2m_max,dewpoint_2m_max,precipitation_sum,rain_sum,snowfall_sum,relative_humidity_2m_mean&models=MPI_ESM1_2_XR"
print(url)

# Send the API request and get the response
response = requests.get(url)
response = response.text
dict_response = json.loads(response)
print(dict_response)

# All responses sorting
# snowfall = dict_response['daily']['snowfall_sum']
# rainsum = dict_response['daily']['rain_sum']
precipiationsum = dict_response['daily']['precipitation_sum']
dewpointmax = dict_response['daily']['dewpoint_2m_max']
relativehummax = dict_response['daily']['relative_humidity_2m_mean']
windspeedmax = dict_response['daily']['windspeed_10m_max']
tempmax = dict_response['daily']['temperature_2m_max']

# snowfall = [x for x in snowfall if x is not None]
# rainsum = [x for x in rainsum if x is not None]
precipiationsum = [x for x in precipiationsum if x is not None]
dewpointmax = [x for x in dewpointmax if x is not None]
relativehummax = [x for x in relativehummax if x is not None]
windspeedmax = [x for x in windspeedmax if x is not None]
tempmax = [x for x in tempmax if x is not None]

# snowfall_max = max(snowfall) if snowfall else None
# rainsum_max = max(rainsum) if rainsum else None
precipiationsum_max = max(precipiationsum) if precipiationsum else None
dewpointmax_max = max(dewpointmax) if dewpointmax else None
relativehummax_max = max(relativehummax) if relativehummax else None
windspeedmax_max = max(windspeedmax) if windspeedmax else None
tempmax_max = max(tempmax) if tempmax else None

# print(f"Maximum snowfall: {snowfall_max}")
# print(f"Maximum rain sum: {rainsum_max}")
print(f"Maximum precipitation sum: {precipiationsum_max}")
print(f"Maximum dew point 2m max: {dewpointmax_max}")
print(f"Maximum relative humidity 2m mean: {relativehummax_max}")
print(f"Maximum wind speed 10m max: {windspeedmax_max}")
print(f"Maximum temperature 2m max: {tempmax_max}")
```
